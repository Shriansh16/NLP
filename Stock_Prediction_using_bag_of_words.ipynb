```python
train = df[df['Date']<'20150101']
test = df[df['Date']>'20141231']
```

```python
# remove punctuation

def remove_punctuation1(sentences):
  result=[]
  for sentence in sentences:
    for i in string.punctuation:
      sentence=sentence.replace(i,'')
    result.append(sentence)
  return result
```

```python
def remove_punctuation2(text):
    if isinstance(text, str):
        for i in string.punctuation:
            text = text.replace(i, '')
    return text
```

```python
data=train.iloc[:,2:27]
```

```python
data.head()
```

```python
data = data.applymap(remove_punctuation2)
```

```python
data
```

```python
# lets rename the column
list=[i for i in range(25)]
new_col=[str(i) for i in list]
```

```python
data.columns=new_col
```

```python
data.head()
```

```python
## lower case
for i in data.columns:
  data[i]=data[i].str.lower()
```

```python
data.head()
```

```python
data.iloc[1,0:25]
```

### now we have to combine them

```python
headlines = []
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[0]
```

```python
# Bag of Words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1500, ngram_range =(2,2))
traindataset = cv.fit_transform(headlines)
```

```python
## for test dataset
test_transform = []
for row in range(0,len(test.index)):
  test_transform.append(' '.join(str(x) for x in test.iloc[row,0:25]))
test_dataset = cv.fit_transform(test_transform)
```

```python
## using random forest classifier
from sklearn.ensemble import RandomForestClassifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
predictions = randomclassifier.predict(test_dataset)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(test['Label'],predictions)
```

```python
print(confusion_matrix(test['Label'],predictions))
```

```python
report=classification_report(test['Label'],predictions)
print(report)
```
