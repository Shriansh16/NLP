```python
from nltk.tokenize import word_tokenize
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
sentence="The Eiffel Tower was built from 1887 to 1889 by Gustave Eiffel's company as the centerpiece of the 1889 World's Fair in Paris, France. Constructed using wrought iron, the tower stands at a height of 324 meters (1,063 feet) and was originally criticized by some as an eyesore. However, over time, it has become an iconic symbol of both Paris and France, attracting millions of visitors annually. Designed by engineers Maurice Koechlin and Émile Nouguier, with architect Stephen Sauvestre contributing the decorative arches at the top, the Eiffel Tower represents a feat of engineering prowess and artistic ingenuity. Today, it remains one of the most recognizable landmarks in the world, showcasing the beauty of both its intricate design and the city it graces."
```

```python
words=word_tokenize(sentence)
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
from nltk import pos_tag,ne_chunk
```

```python
nltk.download('maxent_ne_chunker')
```

```python
nltk.download('words')
```

```python
postags_elements=pos_tag(words)
```

```python
nltk.ne_chunk(postags_elements)
```
